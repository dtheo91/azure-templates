```python
import sys
import os

sys.path.append(os.path.abspath(os.path.join("..")))
from config import config
from modules.azure_foundry.azure_foundry_client import AzureFoundryClient
```

### Get Foundry Client

```python
foundry = AzureFoundryClient(user_endpoint=config.azure_foundry_endpoint, openai_endpoint=config.azure_foundry_openai_endpoint, api_key=config.azure_foundry_api_key)
```

### Get Agent by Name

```python
agent = foundry.project_client.agents.get(agent_name="testagent")
```

### Call Agent

```python
response = foundry.openai_client.responses.create(
    input=[{"role": "user", "content": "Tell me a one line story"}],
    extra_body={"agent": {"name": agent.name, "type": "agent_reference"}},
)

print(f"Response output: {response.output_text}")
```
